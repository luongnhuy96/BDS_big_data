```python
# Hàm chuyển cột 'Price' từ 'object' thành 'float64'
def convertPrice(string):
  res = 0
  pos_ty, pos_trieu, pos_nghin = string.find(" tỷ"), string.find(" triệu"), string.find(" nghìn")
  if pos_ty > 0:
    num = int(string[:pos_ty])
    res+= num*1000000000

    if pos_trieu > 0:
      num = int(string[pos_ty+3:pos_trieu])
      res+= num*1000000 
      if pos_nghin > 0:
        num = int(string[pos_trieu+6:pos_nghin])
        res+= num* 1000 
    else:
      if pos_nghin > 0:
        num = int(string[pos_ty+3:pos_nghin])
        res+= num* 1000000# 
  else:
    if pos_trieu > 0:
      num = int(string[:pos_trieu])
      res+= num*1000000
      if pos_nghin > 0:
        num = int(string[pos_trieu+6:pos_nghin])
        res+= num* 1000
    else:
      if pos_nghin > 0:
        num = int(string[:pos_nghin])
        res+= num*1000
  return res
convertPrice('15 tỷ 500 triệu')
```

```python
def wrangleHCM(data):
  # chuyển Price từ "object" thành "float64" (đơn vị tỷ)
  data['Giá'] = data['Price'].apply(convertPrice)
  # chuyển Using_Square từ "object" thành "float64" (đơn vị m2)
  # data['Diện tích sử dụng'] = data['Using Square'].str.replace(',','.').str[:-3].astype('float')
 
  # chuyển Length, Width từ "object" thành "float64" (đơn vị m)
  data['Dài'] = data['Length'].str.replace(',', '.').astype(float)
  data['Rộng'] = data['Width'].str.replace(',', '.').astype(float)
  # tính lại diện tích = dài x rộng
  data['Diện tích thực (m2)']=data['Dài']*data['Rộng']
 
  # tính giá trên đơn vị m2
  data['Giá/m2']=data['Giá']/data['Diện tích thực (m2)']
    
  # chuyển Post_date từ "object" thành "datetime"
  data['Ngày đăng'] = pd.to_datetime(data['Post Date'], format="%d/%m/%Y")
  
  # bỏ các cột đã xử lý ở trên
  data.drop(columns=['Price', 'Using Square', 'Length', 'Width', 'Post Date', 'Land Square'], inplace=True)
  # Đổi tên cột còn lại thành tiếng Việt
  data.rename(columns = {'Address':'Địa chỉ',
                              'Bed_num': 'Số phòng ngủ',
                              'Bath_num': 'Số phòng tắm',
                              'Legal': 'Pháp lý'},
              inplace = True)

  # thêm cột Loại BĐS
  data['Loại'] = data['Số phòng ngủ'].apply(lambda x: 'CHDV' if x >= 10 else 'Nhà ở')
  # loại bỏ quan trắc trùng lắp
  data.drop_duplicates(inplace=True)
  # loại bỏ BDS với Giá < 500 triệu và > 1000 tỷ (Outlier)
  #data = data.drop([data[(data['Giá'] < 500000000) | (data['Giá'] >= 1000000000000)].index)
  data = data[(data['Giá'] >= 500000000) & (data['Giá'] < 1000000000000)]
  return data
clean_BDS = wrangleHCM(data)
```
